#### Plotting 2 dimensional datasets

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

# method 1

y=np.random.standard_normal((20,2)).cumsum(axis=0)
plt.figure(figsize=(10,6))
plt.plot(y, lw=1.5)
plt.plot(y, 'ro')
plt.xlabel('index')
plt.ylabel('values')
plt.title('A simple plot')
```

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

# method 2

plt.figure(figsize=(10,6))
plt.plot(y[:, 0], lw=1.5, label='1st')
plt.plot(y[:, 1], lw=1.5, label='2nd')
plt.plot(y, 'ro')
plt.legend(loc=0)
plt.xlabel('dates')
plt.ylabel('prices')
plt.title('A range of prices')
```

```python
# loss of information due to scaling of data:

import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

# method 3

y[:, 0] = y[:, 0]*100
plt.figure(figsize=(10,6))
plt.plot(y[:, 0], lw=1.5, label='1st')
plt.plot(y[:, 1], lw=1.5, label='2nd')
plt.plot(y, 'ro')
plt.legend(loc=0)
plt.xlabel('dates')
plt.ylabel('prices')
plt.title('A range of prices')

# the output results in informational loss
```

```python
# the way around dealing with this information loss is:
# a) using two axes
# b) using two subplots
```

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

# method 4 - using two axes

plt.figure(figsize=(10,6))
fig, ax1 = plt.subplots()
plt.plot(y[:, 0], 'b', lw=1.5, label='1st')
plt.plot(y[:, 0], 'ro')
plt.legend(loc=8)
plt.xlabel('dates')
plt.ylabel('oil prices')
plt.title('Oil and LNG prices')
ax2 = ax1.twinx()
plt.plot(y[:, 1], 'g', lw=1.5, label='2nd')
plt.plot(y[:, 1], 'ro')
plt.legend(loc=0)
plt.ylabel('LNG prices')
```

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

# method 5 - using two subplots

plt.figure(figsize=(10,6))
plt.subplot(211)
plt.plot(y[:, 0], 'b', lw=1.5, label='1st')
plt.plot(y[:, 0], 'ro')
plt.legend(loc=0)
plt.ylabel('prices')
plt.title('Oil prices')

plt.subplot(212)
plt.plot(y[:, 1], 'g', lw=1.5, label='2nd')
plt.plot(y[:, 1], 'ro')
plt.legend(loc=0)
plt.ylabel('prices')
plt.xlabel('dates')
plt.title('LNG prices')
```

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

# method 6 - using two different types of subplots (bar and line)

plt.figure(figsize=(10,6))
plt.subplot(121)
plt.plot(y[:, 0], 'b', lw=1.5, label='1st')
plt.plot(y[:, 0], 'ro')
plt.legend(loc=0)
plt.ylabel('prices')
plt.xlabel('dates')
plt.title('Oil prices')

plt.subplot(122)
plt.bar(np.arange(len(y)), y[: , 1], width=0.5, color='g', label='2nd') # this command line creates a bar subplot
plt.legend(loc=0)
plt.xlabel('dates')
plt.title('LNG prices')
```

```python
# creating a scatter plot:

import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

y = np.random.standard_normal((1000,2))
plt.figure(figsize=(10,2))
plt.plot(y[:,0],y[:,1],'ro')
plt.xlabel('age')
plt.ylabel('income')
plt.title('Plot of wealth')
```

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

# method 7 - plotting a histogram:

plt.figure(figsize=(10,6))
plt.hist(y, label=['1st','2nd'], bins=25)
plt.legend(loc=0)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram');
```

```python
# method 8 - stacked histogram:
```

```python
plt.figure(figsize=(10,6))
plt.hist(y, label=['1st','2nd'], color=['b', 'g'], stacked=True, bins=25, alpha=0.5)
plt.legend(loc=0)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram');
```

```python
# method 9 - static 3D plotting of a volatility surface:

# strike values = between 50 and 150
# time to maturity = between 0.5 and 2.5 years
# we first define a 2D ndarray system and then use numpy meshgrid to generate a 3D system

strike = np.linspace(50,150,24)
ttm = np.linspace(0.5,2.5,24)
strike,ttm = np.meshgrid(strike,ttm)
strike[:2].round(1) # output will be an array system generated from 2D ndarray strike and ttm
```

```python
# calculate dummy iv:

iv = (strike-100)**2/(100*strike)/ttm
iv[:5, :3]
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10,6))
ax = fig.gca(projection='3d')
surf = ax.plot_surface(strike,ttm,iv,rstride=2,cstride=2,cmap=plt.cm.coolwarm,linewidth=0.5,antialiased=True)
ax.set_xlabel('strike')
ax.set_ylabel('time-to-maturity')
ax.set_zlabel('implied vol')
fig.colorbar(surf,shrink=0.5,aspect=5)
```
